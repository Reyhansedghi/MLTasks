```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img=cv2.imread('task1.jpg')
```

```python
img=img[:,:,::-1]
plt.imshow(img)
```

```python
img1=img.copy()
img1=img1[:,:,1]
img1=np.where(img1<200,0,255)
img1=img1[100:500,100:1150]
plt.imshow(img1,cmap='gray')
```

```python
#v/h split output is list***
img1v=np.vsplit(img1,2)
num=np.hsplit(img1v[0],5)
num1=np.hsplit(img1v[1],5)
num=num+num1
for i in range(10):
    num[i]=np.array(num[i],dtype=np.uint8)
    num[i] = cv2.resize(num[i], (28, 28)) #cv2 resize func works only with unit8 dtypes***
    plt.imshow(num[i], cmap='gray')
```

```python
for i in range (10):
   num[i] = num[i].flatten()
num[i].shape
```

```python
num=np.asarray(num)
num.shape
```

```python
np.savetxt('num.csv',num,delimiter=',')
```
