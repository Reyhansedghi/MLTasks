```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('mnist_train.csv',header=None)
df
```

```python
Y = df.iloc[:,0]
X = np.array(df.iloc[:,1:])
```

```python
sample_id=np.random.choice(len(X),size=25)
sample_id
```

```python
plt.figure(figsize=(10,10))
for i,idx in enumerate(sample_id):
    plt.subplot(5,5,i+1)
    plt.imshow(X[idx,:].reshape(28,28),cmap='gray',)
    plt.title(Y[idx])
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=.2, stratify= Y, random_state=1)
```

```python
model = LogisticRegression(max_iter=300,random_state=42)
model.fit(x_train, y_train)
```

```python
y_train_pred = model.predict(x_train)
acc_train = accuracy_score(y_train, y_train_pred)
acc_train
```

```python
y_test_pred = model.predict(x_test)
acc_test = accuracy_score(y_test, y_test_pred)
acc_test
```

```python
sample_test_id=np.random.choice(len(x_test),size=25)
sample_test_id
```

```python
plt.figure(figsize=(10,10))
for i,idx in enumerate(sample_test_id):
    plt.subplot(5,5,i+1)
    plt.imshow(x_test[idx,:].reshape(28,28),cmap='gray')
    true_label = y_test.iloc[idx]
    pred_label = y_test_pred[idx]
    plt.title(f"({true_label}, {pred_label})")
    plt.axis('off')
plt.tight_layout()
plt.show()
```
